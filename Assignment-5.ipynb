```python
# Find the factorial function

# Function to calculate factorial

def getFact(num):
    facto = 1
    for i in range(1, num+1):
        facto = facto * i
    return facto


while True:
    
    yourNum = int(input("Enter number to find factorial? "))
    myFact = 1

    if yourNum < 0:
        print('Factorial of negative numbers are not possible.')
        continue
    elif yourNum == 0:
        print('The factorial of 0 is 1.')
        continue
    else:
        print(getFact(yourNum))

    userChoice = input('Do you want more values to calculate factorial? (y/n) ')

    if userChoice == 'y' or userChoice == 'Y':
        continue
    else:
        break
```

```python
# Seperately count Upper, Lower Characters

def sepLowerUpper(myStr):
    
    upperChar  = 0 
    lowerChar = 0
    numChar = 0
    
    for i in myStr:
        if i.isupper():
            upperChar = upperChar + 1
        elif i.isdigit():
            numChar = numChar + 1
        else:
            lowerChar = lowerChar + 1
    return upperChar,lowerChar,numChar

# Collect String text from user

strInput=input('Enter your text...? ')

# Call function to calculate Upper, Lower characters

upperNum, lowerNum, numNum = sepLowerUpper(strInput)


if upperNum <= 1:
    print('There is only ', upperNum, 'upper case character.')
else:
    print('There are ', upperNum, 'uppercase characters in String.')

if lowerNum <= 1:
    print('There is only ', lowerNum, 'lowercase character.')
else:
    print('There are ', lowerNum, 'uppercase characters in String.')

# At the end it will print if any numbers are there
if numNum > 0:
    print('Also there are ', numNum, ' number(s) in this string')
else:
    print('There are no numbers in this text.')

    

```

```python
# Collect even numbers from a list through function

def getEven(list):
    evenList=[]
    for i in list:
        if i % 2 == 0:
            evenList.append(i)
            #print(i)
        else:
            continue
    return evenList


numLimit = int(input('How many numbers you want in list? '))
numList = []

for i in range(1,numLimit+1):
    try:
        numList.append(int(input('Enter number for list : ')))
    except ValueError:
        print('only numbers are requested.')
        numList.append(int(input('Enter number for list : ')))


print('Here is the list', end='')
print(numList)

print('\nNow as list has been Generated, Let\'s seperate the even numbers.\n')

print(getEven(numList))

    

    

```

```python
# Check for Palindrome

# Function to check Palindrome

def checkPalind(inputStr):
    if(inputStr == inputStr[::-1]):
       return True
    else:
        return False

# Collect input

while True:
    
    myStr = input('Enter text to check whether its Palindrome or not : ')

    if checkPalind(myStr):
        print('Yes this word is Palindrome.')
    else:
        print('Sorry this word is "not" Palindrome.')
        print('If you reverse this text it will be ', myStr[::-1])

    cont = input('Want to check more words ? ')

    if(cont == 'y' or cont == 'Y'):
        continue
    else:
        break
    
```

```python
# Check a number if it is prime or not


def isPrime(num):
#
    if num !=2 and num%2 == 0:
        return False
    elif num !=3 and num%3 == 0:
        return False
    elif num != 5 and num%5 == 0:
        return False
    else:
        return True
#

cPrime = int(input('Enter number to check if it is prime : '))

if isPrime(cPrime):
    print('Yes this is a Prime Number.')
else:
    print('Sorry this is not a Prime Number.')
    
```

```python
# Customer Shopping List using arbitrary argument

# Function for collecting shopping items

def buyingList(*items):
    itemsList=[]
    for item in items:
        itemsList.append(item)
    return itemsList

itemsBought=[]

while True:
    boughtItem = input('Enter item you just bought...: ')
    itemsBought.append(boughtItem)

    cont = input('\n\nDo you want to add more itmes? ')
    if cont == 'y' or cont == 'Y':
        continue
    else:
        break

print('\n\nThe items you\'ve bought are : ')
print(buyingList(itemsBought))
```
