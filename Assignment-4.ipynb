```python
"""A movie theater charges different ticket prices depending on a person’s age. If a person is under the age
of 3, the ticket is free; if they are between 3 and 12, the ticket is $10; and if they are over age 12, the
ticket is $15. Write a loop in which you ask users their age, and then tell them the cost of their movie
ticket."""

print('*|* Movie Theatre Ticketing System *|*')

# Variables declarations and initializations 

running = True
totalCostList = []
below3 = 0
between3and12 = 0
above12 = 0

# main Structural loop for the whole program execution

while running:
    custAge = int(input('\nEnter the age of customer : '))

    # checks that in which category of age customer falls
    
    if custAge < 3:
        totalCostList.append(0)
        below3 +=1
        
    elif custAge >=3 and custAge <=12:
        totalCostList.append(10)
        between3and12 +=1
        
    elif custAge > 12:
        totalCostList.append(15)
        above12 +=1
        
    else:
        print('Not a valid age.')

    addMore = input('\nDo you want to buy more tickets?  ')
    if addMore == 'y' or addMore == 'Y':
        continue
    else:
        running = False
        
# On screen printing of age wise sub totals and than total purchase cost

print('You have bought ', below3, ' tickets of under 3 years, i.e. Free')
print('You have bought ', between3and12, ' tickets of ages 3 to 12 years, i.e. $10 x ', between3and12, '= ', between3and12*10)
print('You have bought ', above12, ' tickets of ages 12 and above years, i.e. $15 x ', above12, '= ', above12* 15, '\n')

print('\nYour total bill is : ', sum(totalCostList))
```

```python
"""Guess the number game
Write a program which randomly generate a number between 1 to 30 and ask the user in input field to
guess the correct number. Give three chances to user guess the number and also give hint to user if
hidden number is greater or smaller than the number he given to input field."""

# Call random library

from random import randint

# Generate random number between 1 to 30

mySecret = randint(1,30)

print('Welcome to number guessing game\n\n')

yourGuess = 0

# Main loop to execute 3 tries of guessing the number

for i in range(3):

    if i == 2:
        print('Its your last try dude..!\n')
        
    gs = input('Guess the number - You have total 3 chances only :')
    
    yourGuess = int(gs)

    if yourGuess == mySecret:
        print('Congrats! You win!')
    elif yourGuess > mySecret:
        print('Sorry you are going too high.')
    else:
        print('Sorry your guess is too low.')

    if i == 2:
        print('I am sorry... You loose!')


# Revealing the secret number finally

print('The secret number was : ', mySecret)
```
