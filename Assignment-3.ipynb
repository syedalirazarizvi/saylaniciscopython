```python
# Calculator

#Basic Variable Declaration

result = 0
symb = ""

# Functions

def add(v1, v2):
    return v1+v2

def subt(v1, v2):
    return v1 - v2

def mult(v1, v2):
    return v1 * v2

def divide(v1, v2):
    if v1>0 and v2 > 0:
        return v1 / v2
    else:
        print("Division cannot be performed between 0s")
        symb = '='
        return 0

def modo(v1, v2):
    if v1>0 and v2 > 0:
        return v1 % v2
    else:
        print("Division cannot be performed between 0s")
        symb = '='
        return 0

def power(v1, v2):
    return (v1**v2)


# Start of Main loop

while symb != '=':

# Start of the process by selecting symbol

    symb = input("Select Op symbol (+, -, *, /, mod, ^, =, C) : ")
    if symb == '=' or symb == 'C':
        print('Total = ', result)
        symb = '='

# Addition

    if symb == '+':
        if result != 0:
            value1 = result
            try:
                value2= int(input(str(result)  + symb + " " + " : "))
            except ValueError:
                value2 = 0
                
            result = add(value1, value2)
            print(result)
        else:
            try:
                value1 = int(input("Value 1 : "))
            except ValueError:
                value1 = 0
            try:    
                value2 = int(input("Value 2 : "))
            except ValueError:
                value2 = 0
            result = add(value1, value2)
            print(result)
            

# Subtraction

    if symb == '-':
        if result != 0:
            value1 = result
            try:
                value2= int(input(str(result)  + " " + symb + " : "))
            except ValueError:
                value2 = 0
                
            result = subt(value1, value2)
            print(result)
        else:
            try:
                value1 = int(input("Value 1 : "))
            except ValueError:
                value1 = 0
            try:
                value2 = int(input("Value 2 : "))
            except ValueError:
                value2 = 0
                
            result = subt(value1, value2)
            print(result)

# Multiplication

    if symb == '*':
        if result != 0:
            value1 = result
            try:
                value2= int(input(str(result)  + " " + symb + " : "))
            except ValueError:
                value2 = 0
                
            result = mult(value1, value2)
            print(result)
        else:
            try:
                value1 = int(input("Value 1 : "))
            except ValueError:
                value1 = 0
            try:
                value2 = int(input("Value 2 : "))
            except ValueError:
                value2 = 0
                
            result = mult(value1, value2)
            print(result)

# Division

    if symb == '/':
        try:
            if result != 0:
                value1 = result
                try:
                    value2= int(input(str(result)  + " " + symb + " : "))
                except ValueError:
                    value2 = 0
                    
                result = divide(value1, value2)
                print(result)
            else:
                try:
                    value1 = int(input("Value 1 : "))
                except ValueError:
                    value1 = 0
                try:    
                    value2 = int(input("Value 2 : "))
                except ValueError:
                    value2 = 0
                    
                result = divide(value1, value2)
                print(result)
        except TypeError:
            print('Divide by 0 Error')
            print('Thank you for using simple calc')
            symb = '='
            break
            

# Modulus

    if symb == 'mod':
        try:
            if result != 0:
                value1 = result
                try:
                    value2= int(input(str(result)  + " " + symb + " : "))
                except ValueError:
                    value2 = 0
                    
                result = modo(value1, value2)
                print(result)
            else:
                try:
                    value1 = int(input("Value 1 : "))
                except ValueError:
                    value1 = 0
                try:
                    value2 = int(input("Value 2 : "))
                except ValueError:
                    value2 = 0

                result = modo(value1, value2)
                print(result)
        except TypeError:
            print('Divide by 0 Error')
            print('Thank you for using simple calc')
            symb = '='
            break

# Power

    if symb == '^':
        if result != 0:
            value1 = result
            try:
                value2= int(input(str(result)  + symb + " " + " : "))
            except ValueError:
                value2 = 0
                
            result = add(value1, value2)
            print(result)
        else:
            try:
                value1 = int(input("Value 1 : "))
            except ValueError:
                value1 = 0
            try:    
                value2 = int(input("Value 2 : "))
            except ValueError:
                value2 = 0
            result = power(value1, value2)
            print(result)
            

# Pressing 'C' will clear all variables and Refresh the program

    if symb == 'C' or symb == 'c':
        result = 0
        print(result)
        continue
```

```python
# Finding duplicate values


def checkIfDuplicates_1(myList):
    ''' Check if given list contains any duplicates '''
    if len(myList) == len(set(myList)):
        return False
    else:
        return True


myList = ['Hello', 'Ok', 'is', 'Ok', 'test', 'this', 'is', 'a', 'test']

print('\nCurrent List is : \n')
print(myList)

result = checkIfDuplicates_1(myList)
 
if result:
    print('\nYes, list contains duplicates')
    print('\n', set(myList))
else:
    print('\nNo duplicates found in list')  

```
